## **Challege Exercises**

> 1. Create a DataFrame for storing weather data
> 2. Select a row from mid-point using iloc
> 3. Filter rows matching certain numeric criteria with loc
> 4. Load a CSV file into a DataFrame
> 5. Set the index to another column like dates

```python
# CODE EXAMPLE FOR THE CHALLENGE

import pandas as pd
import matplotlib.pyplot as plt

# Create dummy DataFrame
data = {'name': ['John', 'Mary', 'Sam', 'Sarah'], 
        'height': [183, 167, 174, 160],
        'weight': [89, 56, 68, 50]}
df = pd.DataFrame(data)

# Scatter plot height vs weight
df.plot.scatter(x='height', y='weight')

# Set labels and title
plt.xlabel('Height (cm)') 
plt.ylabel('Weight (kg)')
plt.title('Height vs Weight')

plt.tight_layout()
plt.show()

# Function to filter by height range 
def filter_height(df, min_ht, max_ht):
    return df[(df['height'] >= min_ht) & (df['height'] <= max_ht)]

# Filter rows between 160cm and 170cm height
result = filter_height(df, 160, 170)
print(result)
```

```python
# example of code to use to complet the task
import pandas as pd
import numpy as np  
```

```python
# Sample fruit price DataFrame
data = {'Fruit': ['Apple', 'Banana', 'Orange'],  
        'Price': [2.5, 1.2, 3.3]}
df = pd.DataFrame(data)
```

```python
# Calculate average price  
avg_price = df['Price'].mean()  
print(avg_price)
```

```python
# Filter prices > average
filter = df['Price'] > avg_price
df.loc[filter]
```

```python
# isin() filter on fruits
fruit_filter = df['Fruit'].isin(['Apple', 'Orange'])
df.loc[fruit_filter]
```
